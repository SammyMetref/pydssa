# Example of eSQG reconstruction 
## from eNATL60-no tide SSH map in Western Mediterranean region

```python
import numpy as np
import pylab as plt
import sys, os 
import xarray as xr 
import warnings 
warnings.filterwarnings('ignore')
```

```python
import pydssa.class_vertical_reconstruction as vr
```

## 1. Download SSH field and depth

```python
path_ssh = '../data/Surf_fields/eNATL60MEDWEST-BLB002_y2009m07d01.1h_sossheig.nc'    
with xr.open_dataset(path_ssh, mask_and_scale=True) as (ds_2):   
        day0 = 0 # hour = 0
        crop = 1
        ssh = np.zeros_like(ds_2.sossheig[day0,crop:-crop,crop:-crop]) 
        ssh[:,:] = ds_2.sossheig[day0,crop:-crop,crop:-crop]
        lon = np.zeros_like(ds_2.nav_lon[crop:-crop,crop:-crop]) 
        lon[:,:] = ds_2.nav_lon[crop:-crop,crop:-crop] 
        lat = np.zeros_like(ds_2.nav_lat[crop:-crop,crop:-crop])
        lat[:,:] = ds_2.nav_lat[crop:-crop,crop:-crop]      
        
ssh0 = np.nan_to_num(ssh)
lon1 = lon[lon!=0]
lat1 = lat[lat!=0]
my_lon = np.linspace(np.min(lon1),np.max(lon1),np.shape(lon)[1])
my_lat = np.linspace(np.min(lat1),np.max(lat1),np.shape(lat)[0])
lon0,lat0 = np.meshgrid(my_lon,my_lat)

depth = []
with open('../data/NATL60_depth.txt') as fp:
    for cnt, line in enumerate(fp):
        if cnt != 0 and cnt < 108:
            depth.append(np.float(line.strip()))
             
```

```python
np.shape(ssh0)
```

```python
crop1 = np.argmin((lat0[:,0] - 38)**2)
crop2 = np.argmin((lat0[:,0] - 41)**2)
crop3 = np.argmin((lon0[0,:] - 1)**2)
crop4 = np.argmin((lon0[0,:] - 8)**2)
print(crop1)
print(crop2)
print(crop3)
print(crop4)
```

```python
levs = np.arange(-0.5,0.5, 2*0.6/100) 
plt.contourf(lon0,lat0,ssh0,cmap=plt.cm.get_cmap('bwr'),levels=levs)
#crop1 = 400
#crop2 = -310
#crop3 = 420
#crop4 = -320
plt.plot([lon0[0,crop3],lon0[0,crop3]],[lat0[crop1,:],lat0[crop2,:]],'k',linewidth=2)
plt.plot([lon0[0,crop4],lon0[0,crop4]],[lat0[crop1,:],lat0[crop2,:]],'k',linewidth=2)
plt.plot([lon0[:,crop3],lon0[:,crop4]],[lat0[crop1,0],lat0[crop1,0]],'k',linewidth=2)
plt.plot([lon0[:,crop3],lon0[:,crop4]],[lat0[crop2,0],lat0[crop2,0]],'k',linewidth=2)

#crop1s = np.argmin((lat0[:,0] - 39.7)**2)
#crop2s = np.argmin((lat0[:,0] - 40.7)**2)
#crop3s = np.argmin((lon0[0,:] - 1.5)**2)
#crop4s = np.argmin((lon0[0,:] - 2.5)**2)
#plt.plot([lon0[0,crop3s],lon0[0,crop3s]],[lat0[crop1s,:],lat0[crop2s,:]],'r',linewidth=2)
#plt.plot([lon0[0,crop4s],lon0[0,crop4s]],[lat0[crop1s,:],lat0[crop2s,:]],'r',linewidth=2)
#plt.plot([lon0[:,crop3s],lon0[:,crop4s]],[lat0[crop1s,0],lat0[crop1s,0]],'r',linewidth=2)
#plt.plot([lon0[:,crop3s],lon0[:,crop4s]],[lat0[crop2s,0],lat0[crop2s,0]],'r',linewidth=2)

#plt.plot([lon_ref[0,crop7],lon_ref[0,crop7]],[lat_ref[crop5,:],lat_ref[crop6,:]],'g',linewidth=2)
#plt.plot([lon_ref[0,crop8],lon_ref[0,crop8]],[lat_ref[crop5,:],lat_ref[crop6,:]],'g',linewidth=2)
#plt.plot([lon_ref[:,crop7],lon_ref[:,crop8]],[lat_ref[crop5,0],lat_ref[crop5,0]],'g',linewidth=2)
#plt.plot([lon_ref[:,crop7],lon_ref[:,crop8]],[lat_ref[crop6,0],lat_ref[crop6,0]],'g',linewidth=2)
plt.colorbar()
```

```python
from scipy import interpolate
x = my_lon
y = my_lat
#mask invalid values
array = np.ma.masked_invalid(ssh)
array = np.ma.masked_where(array==0,array)
xx, yy = lon0,lat0 #np.meshgrid(x, y)
#get only the valid values
x1 = xx[~array.mask]
y1 = yy[~array.mask]
newarr = array[~array.mask]

ssh_interp = interpolate.griddata((x1, y1), newarr.ravel(),
                          (xx, yy),
                             method='cubic',fill_value = 0.)
```

```python
lon0 = lon0[crop1:crop2,crop3:crop4]
lat0 = lat0[crop1:crop2,crop3:crop4]
ssh_interp = ssh_interp[crop1:crop2,crop3:crop4]
```

```python
plt.contourf(lon0,lat0,ssh_interp,cmap=plt.cm.get_cmap('bwr'),levels=levs)
plt.colorbar()
```

## 2. Create an ocean reconstruction instance

```python
my_esqg_reconstruction = vr.vertical_reconstruction()
```

## 3. Enter reconstruction parameters

```python
my_esqg_reconstruction.ssh = ssh_interp 
my_esqg_reconstruction.lon = lon0
my_esqg_reconstruction.lat = lat0 
my_esqg_reconstruction.z = depth
f0 = 2*2*np.pi/86164*np.sin(lat0*np.pi/180)
my_esqg_reconstruction.N0 = 50*f0                      
my_esqg_reconstruction.C = 1.5
my_esqg_reconstruction.sppad = 1
my_esqg_reconstruction.Nprofile = np.ones_like(depth) 
my_esqg_reconstruction.periodsmooth = 0
```

## 4. Run the eSQG reconstruction 

```python
my_esqg_reconstruction.run_esqg()
```

```python
levs = np.arange(-0.6,0.6, 2*0.6/100) 
test = np.copy(my_esqg_reconstruction.relvort_norm[:,:,10])
test[ssh_interp==0]=0
test[test>0.6]=0
test[test<-0.6]=0
plt.contourf(lon0,lat0,test,cmap=plt.cm.get_cmap('bwr'),levels=levs)
plt.colorbar()
```

```python
levs = np.arange(-0.001,0.001, 2*0.001/50) 
test = np.copy(my_esqg_reconstruction.buoy[:,:,10])
test[ssh_interp==0]=0
test[test>0.005]=0
test[test<-0.005]=0
plt.contourf(lon0,lat0,test,cmap=plt.cm.get_cmap('bwr'),levels=levs)
plt.colorbar()
```

## 6. Compare reconstructed w to the reference

```python
path_ref = '../data/Reference/eNATL60MEDBAL-BLB002_y2009m07d01.1h_vovecrtz_0-1000m.nc'    
with xr.open_dataset(path_ref, mask_and_scale=True) as (ds_2):    
        w_ref = np.zeros_like(ds_2.vovecrtz[0,:my_esqg_reconstruction.nz,:,:])    
        w_ref[:,:] = ds_2.vovecrtz[0,:my_esqg_reconstruction.nz,:,:] 
        lon_ref = np.zeros_like(ds_2.nav_lon[:,:] )   
        lon_ref[:,:] = ds_2.nav_lon[:,:]   
        lat_ref = np.zeros_like(ds_2.nav_lat[:,:] )   
        lat_ref[:,:] = ds_2.nav_lat[:,:]   
        
w_ref = np.nan_to_num(w_ref)
lon1 = lon_ref[lon_ref!=0]
lat1 = lat_ref[lat_ref!=0]
my_lon = np.linspace(np.min(lon1),np.max(lon1),np.shape(lon_ref)[1])
my_lat = np.linspace(np.min(lat1),np.max(lat1),np.shape(lat_ref)[0])
lon_ref,lat_ref = np.meshgrid(my_lon,my_lat)
        
w_ref = np.ma.masked_invalid(w_ref)
 
```

```python
# Cropping reconstructed w
#crop1 = 197  
#crop2 = 275
#crop3 = 10
#crop4 = 86
crop1 = np.argmin((lat0[:,0] - 39.7)**2)
crop2 = np.argmin((lat0[:,0] - 40.7)**2)
crop3 = np.argmin((lon0[0,:] - 1.5)**2)
crop4 = np.argmin((lon0[0,:] - 2.5)**2)

# Cropping reference w
crop5 = np.argmin((lat_ref[:,0] - lat0[crop1,0])**2)
crop6 = np.argmin((lat_ref[:,0] - lat0[crop2,0])**2)
crop7 = np.argmin((lon_ref[0,:] - lon0[0,crop3])**2)
crop8 = np.argmin((lon_ref[0,:] - lon0[0,crop4])**2)
#crop5 = 0
#crop6 = -1
#crop7 = 0
#crop8 = -1

# Computing levels from w_ref max
#max_var = np.max(np.abs(w_reconst[crop1:crop2,crop3:crop4,10]*1e3)) 
max_var = np.max(np.abs(w_ref[10,crop5:crop6,crop7:crop8]*1e3)) 
levs = np.linspace(-max_var,max_var, 10)
#levs = np.arange(-0.2279,0.2279, 2*0.2279/50) 
w_reconst = np.copy(my_esqg_reconstruction.vertvel[:,:,:])
#w_reconst[ssh_interp==0]=0
#w_reconst[w_reconst>0.25]=0
#w_reconst[w_reconst<-0.25]=0

# Prescribing 3 depths (indices) to plot 
depth0 = 0
depth1 = 40
depth2 = 80

# Prescribing meridional and zonal cross to plot
lon_cross = int(np.shape(my_esqg_reconstruction.vertvel[crop1:crop2,crop3:crop4,depth0])[0]/2)
lat_cross = int(np.shape(my_esqg_reconstruction.vertvel[crop1:crop2,crop3:crop4,depth0])[1]/2)

plt.figure(figsize=(35,6))
plt.subplot(151)
plt.contourf(my_esqg_reconstruction.lon[crop1:crop2,crop3:crop4],my_esqg_reconstruction.lat[crop1:crop2,crop3:crop4],w_reconst[crop1:crop2,crop3:crop4,depth0]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar() 
plt.ylabel('eSQG',fontsize=18)
plt.title('Vertical velocity (10$^{-3}$m/s) at '+str("%.1f" %my_esqg_reconstruction.z[depth0])+' m',fontsize=15)
plt.subplot(152) 
#levs = np.linspace(-max_var,max_var, 10)
plt.contourf(my_esqg_reconstruction.lon[crop1:crop2,crop3:crop4],my_esqg_reconstruction.lat[crop1:crop2,crop3:crop4],w_reconst[crop1:crop2,crop3:crop4,depth1]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar()
plt.title('Vertical velocity (10$^{-3}$m/s) at '+str("%.1f" %my_esqg_reconstruction.z[depth1])+'m',fontsize=15)
plt.subplot(153) 
#levs = np.linspace(-max_var,max_var, 10)
plt.contourf(my_esqg_reconstruction.lon[crop1:crop2,crop3:crop4],my_esqg_reconstruction.lat[crop1:crop2,crop3:crop4],w_reconst[crop1:crop2,crop3:crop4,depth2]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar() 
plt.title('Vertical velocity (10$^{-3}$m/s) at '+str("%.1f" %my_esqg_reconstruction.z[depth2])+'m',fontsize=15)
plt.subplot(154)
plt.contourf(my_esqg_reconstruction.lon[lat_cross,crop3:crop4],my_esqg_reconstruction.z[:],np.transpose(w_reconst[crop1+lat_cross,crop3:crop4,:]*1e3),cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar()
plt.title('Zonal cross section at '+str("%.1f" % my_esqg_reconstruction.lat[crop1+lat_cross,1])+'N (10$^{-3}$m/s)',fontsize=15)
plt.subplot(155)
plt.contourf(my_esqg_reconstruction.lat[crop1:crop2,lon_cross],my_esqg_reconstruction.z[:],np.transpose(w_reconst[crop1:crop2,crop3+lon_cross,:]*1e3),cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar()
plt.title('Meridional cross section at '+str("%.1f" % np.abs(my_esqg_reconstruction.lon[1,crop3+lon_cross]))+'W (10$^{-3}$m/s)',fontsize=15)
  
# Computing levels from w_ref max
max_var = np.max(np.abs(w_ref[10,crop5:crop6,crop7:crop8]*1e3)) 
levs = np.linspace(-max_var,max_var, 10)

plt.figure(figsize=(35,6))
plt.subplot(151)
plt.contourf(lon_ref[crop5:crop6,crop7:crop8],lat_ref[crop5:crop6,crop7:crop8],w_ref[depth0,crop5:crop6,crop7:crop8]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar() 
plt.ylabel('Reference',fontsize=18)
plt.title('Vertical velocity (10$^{-3}$m/s) at '+str("%.1f" %my_esqg_reconstruction.z[depth0])+' m',fontsize=15)
plt.subplot(152)
plt.contourf(lon_ref[crop5:crop6,crop7:crop8],lat_ref[crop5:crop6,crop7:crop8],w_ref[depth1,crop5:crop6,crop7:crop8]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar()
plt.title('Vertical velocity (10$^{-3}$m/s) at '+str("%.1f" %my_esqg_reconstruction.z[depth1])+'m',fontsize=15)
plt.subplot(153)
plt.contourf(lon_ref[crop5:crop6,crop7:crop8],lat_ref[crop5:crop6,crop7:crop8],w_ref[depth2,crop5:crop6,crop7:crop8]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar() 
plt.title('Vertical velocity (10$^{-3}$m/s) at '+str("%.1f" %my_esqg_reconstruction.z[depth2])+'m',fontsize=15)
plt.subplot(154)
plt.contourf(lon_ref[lat_cross,crop7:crop8],my_esqg_reconstruction.z[:], w_ref[:,lat_cross,crop7:crop8]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar()
plt.title('Zonal cross section at '+str("%.1f" % lat_ref[lat_cross,1])+'N (10$^{-3}$m/s)',fontsize=15)
plt.subplot(155)
plt.contourf(lat_ref[crop5:crop6,lon_cross],my_esqg_reconstruction.z[:],w_ref[:,crop5:crop6,lon_cross]*1e3,cmap=plt.cm.get_cmap('bwr'),levels=levs,extend='both')
plt.colorbar()
plt.title('Meridional cross section at '+str("%.1f" % np.abs(lon_ref[1,lon_cross]) )+'W (10$^{-3}$m/s)',fontsize=15)
plt.show()
```
